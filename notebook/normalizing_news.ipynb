```python
from nlputils import lexical
lexical_normalizer = lexical.Preprocessing()
```

```python
import os
corpora_path = '../data/corpora/CSTNews 6.0'
corpora_dirs = os.listdir(corpora_path)
corpora_dirs = [d for d in corpora_dirs if d not in '.DS_Store']
corpora_dirs.remove('For all the clusters')
```

```python
news = {}
for news_info in corpora_dirs:
    noticias = os.listdir(corpora_path + '/' + news_info + '/Textos-fonte')
    if news_info.split('_')[1] not in news:
        news[news_info.split('_')[1]] = {'Source': [], 'Name': [], 'Text': []}
    for file_title in noticias:
        if '_titulo' not in file_title:
            info = file_title.split('_')
            news[news_info.split('_')[1]]['Source'].append(info[2].replace('.txt', ''))
            news[news_info.split('_')[1]]['Name'].append(news_info.split('_')[2])
            file = corpora_path + '/' + news_info + '/Textos-fonte/' + file_title
            with open(file, 'r', encoding='utf-8') as text_file:
                lines = text_file.readlines()
                text = []
                for line in lines:
                    if line != '\n':
                        line = line.replace('\n','')
                        text.append(line)
                news[news_info.split('_')[1]]['Text'].append(text)
```

```python
news.keys()
```

```python
for theme in news:
    news[theme]['Text'] = [lexical_normalizer.remove_punctuation(x) for x in news[theme]['Text']]
    news[theme]['Text'] = [lexical_normalizer.lowercase(x) for x in news[theme]['Text']]
    news[theme]['Text'] = [lexical_normalizer.tokenize_words(x) for x in news[theme]['Text']]
    news[theme]['Text'] = [lexical_normalizer.remove_stopwords(x) for x in news[theme]['Text']]
```
