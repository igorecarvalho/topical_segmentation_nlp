```python
#importação das bibliotecas criadas para normalização dos dados
from nlputils import lexical
#from nlputils import morphosyntax
#from nlputils import syntax
from nlputils import read_data
```

```python
lexical_normalizer = lexical.Preprocessing()
#morphosyntax_normalizer = morphosyntax.Preprocessing()
#syntax_normalizer = syntax.Preprocessing()
corpora = read_data.Preprocessing()
```

```python
path = '../data/Topicos'
news_xls = corpora.path_news(path)
```

```python
df_news, dupl_news = corpora.dataFrame_news(news_xls)
```

```python
match = corpora.news_match(df_news, dupl_news)
match
```

```python
df_news['C42.2']['Texto'] = df_news['C42.2']['Texto'].apply(lexical_normalizer.tokenize_sentences)
df_news['C42.2']['Texto'] = df_news['C42.2']['Texto'].apply(lexical_normalizer.remove_punctuation)
df_news['C42.2']['Texto'] = df_news['C42.2']['Texto'].apply(lexical_normalizer.lowercase)
df_news['C42.2']['Texto'] = df_news['C42.2']['Texto'].apply(lexical_normalizer.tokenize_words)
df_news['C42.2']['Texto'] = df_news['C42.2']['Texto'].apply(lexical_normalizer.remove_stopwords)
```

```python
df_news['C11.3']['Tópico (Rotule apenas Inicio)'] = df_news['C11.3']['Tópico (Rotule apenas Inicio)'].apply(lexical_normalizer.tokenize_sentences)
df_news['C11.3']['Tópico (Rotule apenas Inicio)'] = df_news['C11.3']['Tópico (Rotule apenas Inicio)'].apply(lexical_normalizer.remove_punctuation)
df_news['C11.3']['Tópico (Rotule apenas Inicio)'] = df_news['C11.3']['Tópico (Rotule apenas Inicio)'].apply(lexical_normalizer.lowercase)
df_news['C11.3']['Tópico (Rotule apenas Inicio)'] = df_news['C11.3']['Tópico (Rotule apenas Inicio)'].apply(lexical_normalizer.tokenize_words)
df_news['C11.3']['Tópico (Rotule apenas Inicio)'] = df_news['C11.3']['Tópico (Rotule apenas Inicio)'].apply(lexical_normalizer.remove_stopwords)
```
